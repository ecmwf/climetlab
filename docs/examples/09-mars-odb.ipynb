# ECMWF Observation bespoke format (ODB)

```python
import climetlab as cml
```

```python
source = cml.load_source(
    "mars",
    type="ofb",
    obsgroup="conv",
    time="12",
    format="odb",
    reportype=16001,
    date="2020-05-18",
)
```

```python
pd = source.to_pandas()
```

```python
cml.plot_map(pd)
```
