# Meteonet

This is an experiment to plot a radar image from https://github.com/meteofrance/meteonet

```python
import climetlab as cml
```

## Radar images

```python
ds = cml.load_dataset("meteonet-radar")
```

```python
ds.home_page
```

```python
ds.licence
```

```python
cml.plot_map(ds, grid=True)
```

```python
ds.to_xarray()
```

## Ground station observations

```python
ds = cml.load_dataset("meteonet-ground-stations", domain='SE')
```

```python
ds.to_pandas()
```

```python
cml.plot_map(ds)
```
